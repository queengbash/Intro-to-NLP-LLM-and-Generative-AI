# Mini NLP Project: Rule-Based FAQ Chatbot

This notebook builds a simple chatbot that answers a few common questions using keyword matching. It demonstrates core NLP ideas such as text normalization, dictionary lookup, and rule-based response selection.

## Project Goal

We want a chatbot that can respond to a small set of user questions without using any machine learning model. Instead, the program will:
- normalize the user's text to lowercase
- look for important keywords
- return a matching predefined answer

```python
# Project: Mini FAQ Chatbot
# Created with Copilot assistance
# Learned: text normalization, keyword matching, dictionary lookup

import re

# FAQ knowledge base: keyword -> reply
faq = {
    "hours": "We are open from 9:00 AM to 5:00 PM, Monday through Friday.",
    "deadline": "The assignment deadline is July 15, 2026 at 11:59 PM EDT.",
    "location": "The workshop is held online through the course portal.",
    "instructor": "The instructor is Joseph Luper Tsenum.",
    "contact": "You can contact the organizers at contact@neliref.org."
}

def normalize_text(text):
    """Convert input to lowercase and remove punctuation for consistent matching."""
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    return text.strip()

def chatbot_reply(user_input):
    """Return a canned reply based on keyword matching."""
    cleaned_text = normalize_text(user_input)
    
    # Check the most important keywords first
    for keyword, reply in faq.items():
        if keyword in cleaned_text:
            return reply
    
    return "I'm not sure. Please ask a question about hours, deadline, location, instructor, or contact."

# Demo the chatbot with a few example questions
sample_questions = [
    "What are your hours?",
    "When is the deadline?",
    "Where is the workshop?",
    "Who is the instructor?",
    "How can I contact you?",
    "Tell me about the weather"
]

for question in sample_questions:
    print(f"Q: {question}")
    print(f"A: {chatbot_reply(question)}\n")
```

## How the Chatbot Works

1. The user input is converted to lowercase and punctuation is removed.
2. The cleaned text is checked against the FAQ keywords.
3. If a keyword is found, a predefined response is returned.
4. If no keyword matches, the chatbot gives a fallback message.
5. You can now type new questions in the interactive demo cell to test fresh inputs.

```python
# Interactive demo: ask the chatbot a question
print("Type 'quit' or 'exit' to leave the chat.\n")
while True:
    user_input = input("You: ")
    if user_input.strip().lower() in {"quit", "exit"}:
        print("Chatbot: Goodbye!")
        break
    print(f"Chatbot: {chatbot_reply(user_input)}\n")
```
